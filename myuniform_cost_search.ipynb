```python
import heapq
def uniform_cost_search(start, goal, graph, cost):
    # Initialize priority queue 
    priority_queue = [(0, start)]
    # Initialize 'dictionary' to track cost and path to each node
    answer = {node: {'cost': float('inf'), 'path': []} for node in range(len(graph))}
    # Set cost of start node to 0
    answer[start]['cost'] = 0

    # While priority queue isn't empty
    while priority_queue:
        # Pop node with lowest cost from queue
        current_cost, current_node = heapq.heappop(priority_queue)
        
        # If current node is goal node, return minimum cost and path
    if current_node in goal:
     print("Goal node reached:", current_node)
     print("Value of answer:", answer)
     return answer
        
        # Check neighbors of current node
     for neighbor in graph[current_node]:
            # Calculate cost to reach neighbor
         total_cost = current_cost + cost[(current_node, neighbor)]
            # If total cost is less than current cost to reach neighbor
     if total_cost < answer[neighbor]['cost']:
                # Update cost and path to neighbor
        answer[neighbor]['cost'] = total_cost
     answer[neighbor]['path'] = answer[current_node]['path'] + [current_node]
                # Add neighbor to priority queue with its new cost
     heapq.heappush(priority_queue, (total_cost, neighbor))
    
 
    

# main function
if __name__ == '__main__':
     
    # create a graph with no more than 30 nodes
    graph, cost = [[] for i in range(30)], {}
 
    # add edges to the graph
    graph[0].append(4)
    graph[0].append(5)
    graph[0].append(16)
    graph[2].append(1)
    graph[3].append(1)
    graph[4].append(2)
    graph[4].append(3)
    graph[4].append(5)
    graph[5].append(8)
    graph[5].append(18)
    graph[6].append(3)
    graph[6].append(7)
    graph[8].append(16)
    graph[8].append(17)
    graph[16].append(17)
    graph[18].append(6)
    
 
    # add cost to each edge
    cost[(0, 4)] = 3
    cost[(0, 5)] = 9
    cost[(0, 16)] = 1
    cost[(2, 1)] = 2
    cost[(3, 1)] = 2
    cost[(4, 2)] = 1
    cost[(4, 3)] = 8
    cost[(4, 5)] = 2
    cost[(5, 8)] = 3
    cost[(5, 18)] = 2
    cost[(6, 3)] = 3
    cost[(6, 7)] = 2
    cost[(8, 16)] = 4
    cost[(8, 17)] = 4
    cost[(16, 17)] = 15
    cost[(18, 6)] = 1
    
    # set start state 
    start = 0
    
    # set goal state, there can be multiple goal states
    goal = [7]
    
    # call uniform_search_cost function to get the minimum cost to reach gaol. Bonus points for path
   
    min_cost_info = uniform_cost_search(start, goal, graph, cost)
    print("Result of uniform_cost_search:", min_cost_info)

    for node, info in min_cost_info.items():
        print(f'Minimum cost from {start} to {node} is {info["cost"]}')
      
    #   **** Bonus ****
    #   print(f'Path: {info["path"]}')
```
